```python
# Basic data science packages
import pandas as pd

# For visualization
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'


# Import raw data
# Read file from current directory
df = pd.read_csv('/Users/sarahamiraslani/Desktop/BH_straight_from_Qualtrics+coding.csv', header = 0)

# Drop the extra header exported by qualtrics
df = df.drop([0,1],axis =0)
```

```python
# Wordcloud
txt =''.join(major for major in df.Major)

# Stop words i.e., strings that will not be included in word cloud
stopwords = set(STOPWORDS)
stopwords.update(['and','major','minor','minors','minoring in',
                  '(major)', '(minor)','BS','B.S.','&',
                  '-', 'with a specialization in','emphasis', 'spec', 
                 'specialization', 'specialized', 'Minor', 'Major',
                  'minorMajor','double','Sci'])

wordcloud = WordCloud(width=800, height=400, stopwords = stopwords, background_color = 'white').generate(txt)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
